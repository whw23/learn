```python
from duckduckgo_search import DDGS
import json
 
results = DDGS().text("deepseek",region="cn-zh",max_results=4)  # 搜索北京
 
# 将results转为json格式
json_results = json.dumps(results, ensure_ascii=False, indent=4)
print(json_results)
```

```python
from baidusearch.baidusearch import search

baidu_results = search('China', num_results=20)  # 返回20条或更少的结果

json_baidu_results=json.dumps(baidu_results, ensure_ascii=False, indent=4)
print(json_baidu_results)
```
